# Derivative 

```python
from scipy.integrate import odeint
import numpy as np
import pylab
from math import asin
```

```python
def finite_diff(f,x,h=0.0001):
    df =(f(x+h) - f(x-h))/(2.0*h)
    ddf =(f(x+h) - 2.0*f(x) + f(x-h))/h**2
    return df,ddf
x = 0.99
df,ddf = finite_diff(asin,x)
print ('First derivative =',df)
print ('Second derivative =',ddf)
```

```python
import numpy as np
import astropy.units as unit
```

```python
H0 = 67.7*unit.km/unit.s/unit.Mpc
print("H0 = {:.2e}".format(H0.to(1/unit.Gyr)))
# dictionary of cosmological models
cosmology = {'standard' : (0.309, 1e-5, 0.691), 'matter sub' : (0.309, 0, 0), 'matter crit' : (1, 0, 0), 'matter super' : (2, 0, 0),}
```

```python
def dota(t, a, OmegaM, OmegaR, OmegaL, H0):
    Omega0 = OmegaM + OmegaR + OmegaL
    return H0 * (OmegaM/a + OmegaR/a**2 + OmegaL*a**2 + 1 - Omega0)**(1/2)
```

```python
import numpy as np
from scipy.constants import G,year,au
from astropy.constants import M_sun
M1 = 2.06*M_sun.value # mass of Sirius A
M2 = 1.02*M_sun.value # mass of Sirius B
a = 2.64*7.4957*au # semi-major axis
e = 0.5914
```

```python
T = 2*np.pi * (G*(M1 + M2))**(-1/2) * a**(3/2)
print("Orbital period = {:.1f} yr".format(T/year))
```

```python
from scipy import integrate
H0_num = H0.to(1/unit.Gyr).value
OmegaM, OmegaR, OmegaL = cosmology['matter super']
Omega0 = OmegaM + OmegaR + OmegaL
tmp = integrate.quad(
lambda a: (OmegaM/a + OmegaR/a**2 + OmegaL*a**2 +
1 - Omega0)**(-1/2),0, 2)
t_crunch = 2*tmp[0]/H0_num
print("Time of Big Crunch: {:.1f} Gyr".format(t_crunch))
```
