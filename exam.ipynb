```python
import numpy as np
import pylab
from numpy import pi
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,10*pi,100)
f=np.sin(x)
g=np.cos(x)
pylab.figure(figsize=(14,10))
pylab.plot(x,f,'r',label='sinx',linewidth=4)
pylab.plot(x,g,'b',label='cos',linewidth=4)
pylab.xlabel('x axis',fontsize=20)
pylab.ylabel('sinx and cosx ',fontsize=20)
pylab.legend(loc=4)
pylab.show()
```

```python
m=np.linspace(0.98,1.8,100)
lo=1.0
mo=1.99e+30 # mass of the sun

l= (m/mo)*lo**2

pylab.figure(figsize=(8,6))
pylab.plot(m,l,label='m vs l',linewidth=5)
pylab.xlabel('mass')
pylab.ylabel('l')
pylab.legend(loc=3)
pylab.show()
```

```python
from scipy.integrate import odeint
```

```python
def mode (y,x):
    k=4.0
    dydx= 4 * k * x
    return dydx
yo=4

x=np.linspace(1,2*pi,100)

y=odeint(mode,yo,x)

pylab.figure(figsize=(8,6))
pylab.plot(x,y,label='m vs l')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.legend(loc=3)
pylab.show()
    
    
    
```
