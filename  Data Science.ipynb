```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
# bars are by default width 0.8, so we'll add 0.1 to the left coordinates
# so that each bar is centered
xs = [i + 0.1 for i, _ in enumerate(movies)]
# plot bars with left x-coordinates [xs], heights [num_oscars]
plt.figure(figsize=(14,10))
plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards",fontsize=27)
plt.title("My Favorite Movies",fontsize=27)
# label x-axis with movie names at bar centers
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]
plt.figure(figsize=(14,6))
plt.bar([2012.6, 2013.6], mentions, 0.6)

plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'",fontsize=17)
plt.ticklabel_format(useOffset=False)
# misleading y-axis only shows the part above 500

plt.axis([2012.5,2014.5,499,506])
plt.title("Look at the 'Huge' Increase!",fontsize=17)

plt.show()
```

```python
variance =[1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared =[256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error =[x + y for x, y in zip(variance, bias_squared)]
xs = [i for i,_ in enumerate(variance)]
plt.figure(figsize=(14,8))
plt.plot(xs, variance,'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')
plt.legend(loc=9)
plt.xlabel("model complexity",fontsize=17)
plt.title("The Bias-Variance Tradeoff",fontsize=17)
plt.legend(loc=9, fontsize =14)
plt.show()
```

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.figure(figsize=(14,8))
plt.scatter(friends, minutes)
# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,xy=(friend_count, minute_count),xytext=(5, -5),textcoords='offset points')
plt.title("Daily Minutes vs. Number of Friends",fontsize=20)
plt.xlabel("# of friends",fontsize=20)
plt.ylabel("daily minutes spent on the site",fontsize=20)
plt.show()
```
