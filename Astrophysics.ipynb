```python
import numpy as np
import pylab
from math import pi
```

```python
x=np.linspace(60,2 * pi,1000)
y=np.sin(x)
pylab.figure(figsize=(16,10))
pylab.plot(x,y,'g',label='sin(x)',linewidth=6)
pylab.xlabel('X axis',fontsize=25)
pylab.ylabel('sine',fontsize=25)
pylab.title('sine function ',fontsize=25)
pylab.legend(loc=4,fontsize=25)
pylab.show()
```

```python
x=np.linspace(60,2 * pi,1000)
y=np.cos(x)
pylab.figure(figsize=(16,10))
pylab.plot(x,y,'b',label='cos(x)',linewidth=6)
pylab.xlabel('X axis',fontsize=25)
pylab.ylabel('cosine',fontsize=25)
pylab.title('cosine function ',fontsize=25)
pylab.legend(loc=4,fontsize=25)
pylab.show()
```
