## Simple Python Code

```python
import numpy as np
import pylab
from numpy import pi
```

```python
x= np.linspace(20 ,2 * pi,1000)
f= np.sin(x)
pylab.figure(figsize=(16,10))
pylab.plot(x,f,'b',label='sin(x)',linewidth=5)
pylab.xlabel('x axis',fontsize=20)
pylab.ylabel('sine function',fontsize=20)
pylab.title('graph of  function ',fontsize=20)
pylab.legend(loc=4,fontsize=20)
pylab.show()
```

```python
x=np.linspace(20,2 * pi ,100)
f=np.cos(x)
pylab.figure(figsize=(12,10))

pylab.plot(x,f,'g',label='cos(x)',linewidth=3)
pylab.xlabel('x axis',fontsize=20)
pylab.ylabel('cosine function',fontsize=20)
pylab.title('graph of  function ',fontsize=20)
pylab.legend(loc=4,fontsize=20)
pylab.show()

```

```python
x= np.linspace(1, 8 * pi,100)
f=np.tan(x)
pylab.figure(figsize=(14,8))
pylab.xlim(5, 15)
pylab.ylim(-1.2, 1.2)
pylab.plot(x,f,'g',label='tanx',linewidth=5)
pylab.xlabel('x axis',fontsize=20)
pylab.ylabel('Tangent function',fontsize=20)
pylab.title('graph of  function ',fontsize=20)
pylab.legend(loc=4,fontsize=20)
pylab.show()

```

```python
x=np.linspace(-100, 100 ,100)
f=  x**2
pylab.figure(figsize=(12,10))

pylab.plot(x,f,'g',label='$x^{2}$',linewidth=3)
pylab.xlabel('x axis',fontsize=20)
pylab.ylabel('Qudaratic function',fontsize=20)
pylab.title('graph of  function ',fontsize=20)
pylab.legend(loc=4,fontsize=20)
pylab.show()
```
